```python
# Projet image : detection des pieces monnai

import math
import numpy as np
import cv2
import matplotlib.pyplot as plt
from collections import deque

def show(img, title="", cmap=None, figsize=(6,6)):
    plt.figure(figsize=figsize)
    if img is None:
        plt.title(title + " (None)")
        plt.axis("off")
        plt.show()
        return
    if img.ndim == 2:
        plt.imshow(img, cmap=cmap or "gray")
    else:
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis("off")
    plt.show()

def euros_str(total):
    euros = int(total)
    cents = int(round((total - euros) * 100))
    if cents == 100:
        euros += 1
        cents = 0
    return f"{euros}€ et {cents:02d} centimes"

# Otsu threshold

def otsu_threshold(image_u8):
    hist, _ = np.histogram(image_u8.flatten(), bins=256, range=(0,256))
    total = image_u8.size
    sum_total = np.dot(np.arange(256), hist)

    sum_bg = 0.0
    w_bg = 0.0
    max_var = -1.0
    best_t = 0

    for t in range(256):
        w_bg += hist[t]
        if w_bg == 0:
            continue
        w_fg = total - w_bg
        if w_fg == 0:
            break

        sum_bg += t * hist[t]
        mu_bg = sum_bg / w_bg
        mu_fg = (sum_total - sum_bg) / w_fg

        var_between = w_bg * w_fg * (mu_bg - mu_fg) ** 2
        if var_between > max_var:
            max_var = var_between
            best_t = t
    return best_t

def otsu_binarize(gray_u8):
    t = otsu_threshold(gray_u8)
    bin_img = (gray_u8 > t).astype(np.uint8)  # 0/1
    return bin_img, t

# morphology

def morph_cleanup(binary01, close_ksize=11, open_ksize=5):
    """
    binary01: 0/1 uint8 mask
    returns: cleaned 0/1 uint8 mask
    """
    x = (binary01 * 255).astype(np.uint8)
    k_close = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (close_ksize, close_ksize))
    k_open  = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (open_ksize, open_ksize))
    x = cv2.morphologyEx(x, cv2.MORPH_CLOSE, k_close)
    x = cv2.morphologyEx(x, cv2.MORPH_OPEN,  k_open)
    return (x > 0).astype(np.uint8)

# Connected components (4-neighbour)

def connected_components_4(binary):
    h, w = binary.shape
    labels = np.zeros((h, w), dtype=np.int32)
    current = 1

    for y in range(h):
        for x in range(w):
            if binary[y, x] == 1 and labels[y, x] == 0:
                q = deque([(y, x)])
                labels[y, x] = current
                while q:
                    cy, cx = q.popleft()
                    for dy, dx in [(-1,0),(1,0),(0,-1),(0,1)]:
                        ny, nx = cy + dy, cx + dx
                        if 0 <= ny < h and 0 <= nx < w:
                            if binary[ny, nx] == 1 and labels[ny, nx] == 0:
                                labels[ny, nx] = current
                                q.append((ny, nx))
                current += 1
    return labels, current - 1

def component_stats(labels, n_labels):
    comps = []
    for lab in range(1, n_labels + 1):
        ys, xs = np.where(labels == lab)
        if xs.size == 0:
            continue
        area = int(xs.size)
        x0, x1 = int(xs.min()), int(xs.max())
        y0, y1 = int(ys.min()), int(ys.max())
        cx = float(xs.mean())
        cy = float(ys.mean())
        comps.append({
            "label": lab,
            "area": area,
            "bbox": (x0, y0, x1, y1),
            "centroid": (cx, cy)
        })
    return comps

# Radius from area

def radius_from_area(area):
    return math.sqrt(area / math.pi)

# Color features

def ring_center_features_hsv(bgr, cx, cy, r):
    """
    Measure HSV in two zones:
    - inner disk (center)
    - outer ring (annulus)
    Returns: (Hc, Sc, Vc, Hr, Sr, Vr, deltaH, deltaS)
    """
    h, w = bgr.shape[:2]
    hsv = cv2.cvtColor(bgr, cv2.COLOR_BGR2HSV)

    mask_center = np.zeros((h, w), dtype=np.uint8)
    mask_ring   = np.zeros((h, w), dtype=np.uint8)

    cx_i, cy_i = int(round(cx)), int(round(cy))
    r_center = max(1, int(round(0.45 * r)))
    r_in     = max(1, int(round(0.70 * r)))
    r_out    = max(1, int(round(0.95 * r)))

    cv2.circle(mask_center, (cx_i, cy_i), r_center, 255, -1)
    cv2.circle(mask_ring,   (cx_i, cy_i), r_out,    255, -1)
    cv2.circle(mask_ring,   (cx_i, cy_i), r_in,     0,   -1)

    Hc, Sc, Vc, _ = cv2.mean(hsv, mask=mask_center)
    Hr, Sr, Vr, _ = cv2.mean(hsv, mask=mask_ring)

    deltaH = abs(float(Hc) - float(Hr))
    deltaS = abs(float(Sc) - float(Sr))

    return float(Hc), float(Sc), float(Vc), float(Hr), float(Sr), float(Vr), float(deltaH), float(deltaS)

def classify_family_by_ring_center(Hc, Sc, Hr, Sr, deltaH, deltaS,
                                  bimetal_deltaH=12.0,
                                  bimetal_deltaS=35.0,
                                  copper_H_max=18.0):
    """
    Baseline rules (tune slightly with your dataset):
    - bimetal if center vs ring differs enough
    - else copper vs gold by hue
    """
    if (deltaH >= bimetal_deltaH) or (deltaS >= bimetal_deltaS):
        return "bimetal"
    if Hc <= copper_H_max:
        return "copper"
    return "gold"

# Normalized radius classification (robust to camera distance)

EURO_DIAM_MM = {
    2.00: 25.75,
    1.00: 23.25,
    0.50: 24.25,
    0.20: 22.25,
    0.10: 19.75,
    0.05: 21.25,
    0.02: 18.75,
    0.01: 16.25,
}
EURO_RATIO = {v: d / EURO_DIAM_MM[2.00] for v, d in EURO_DIAM_MM.items()}

FAMILY_VALUES = {
    "bimetal": [1.00, 2.00],
    "gold":    [0.10, 0.20, 0.50],
    "copper":  [0.01, 0.02, 0.05],
}

def classify_by_normalized_radius(family, r, r_max):
    """
    r_max = largest radius in the image (assumed scale reference ~ 2€)
    rn = r / r_max; choose the closest known diameter ratio within that family.
    """
    if r_max <= 1e-6:
        return None
    rn = r / r_max
    candidates = FAMILY_VALUES[family]
    best_val = candidates[0]
    best_err = 1e9
    for val in candidates:
        err = abs(rn - EURO_RATIO[val])
        if err < best_err:
            best_err = err
            best_val = val
    return float(best_val)

# -------------------------
# Main pipeline
# -------------------------
def detect_and_count_coins(
    bgr,
    max_size=900,
    blur_ksize=5,
    close_ksize=11,
    open_ksize=5,
    min_area=1500,
    debug=True
):
    # 1) Resize
    h0, w0 = bgr.shape[:2]
    scale = max_size / max(h0, w0)
    if scale < 1.0:
        bgr = cv2.resize(bgr, (int(w0*scale), int(h0*scale)), interpolation=cv2.INTER_AREA)

    # 2) Preprocess
    gray = cv2.cvtColor(bgr, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray, (blur_ksize, blur_ksize), 0)

    # 3) Otsu + invert if needed
    bin_img, t = otsu_binarize(blur)
    if bin_img.mean() > 0.5:
        bin_img = 1 - bin_img

    # 4) Morphology cleanup
    cleaned = morph_cleanup(bin_img, close_ksize=close_ksize, open_ksize=open_ksize)

    # 5) Connected components + stats
    labels, n = connected_components_4(cleaned)
    comps = component_stats(labels, n)

    # 6) Filter by area and compute radius
    kept = []
    for c in comps:
        if c["area"] < min_area:
            continue
        c["radius"] = radius_from_area(c["area"])
        kept.append(c)

    # Need r_max for normalization
    r_max = max([c["radius"] for c in kept], default=0.0)

    # 7) Classify each coin: family (color) + normalized radius (value)
    for c in kept:
        cx, cy = c["centroid"]
        r = c["radius"]
        Hc, Sc, Vc, Hr, Sr, Vr, dH, dS = ring_center_features_hsv(bgr, cx, cy, r)
        fam = classify_family_by_ring_center(Hc, Sc, Hr, Sr, dH, dS)
        val = classify_by_normalized_radius(fam, r, r_max)

        c["family"] = fam
        c["value"] = val
        c["Hc"], c["Sc"], c["Hr"], c["Sr"] = Hc, Sc, Hr, Sr
        c["dH"], c["dS"] = dH, dS

    values = [c["value"] for c in kept if c["value"] is not None]
    total = float(np.sum(values)) if values else 0.0

    # 8) Overlay for visualization
    overlay = bgr.copy()
    for c in kept:
        cx, cy = c["centroid"]
        r = c["radius"]
        val = c["value"]
        fam = c["family"]
        cv2.circle(overlay, (int(round(cx)), int(round(cy))), int(round(r)), (255,0,0), 2)
        cv2.putText(overlay, f"{fam} {val:.2f}€",
                    (int(cx)+5, int(cy)-5),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,255), 1, cv2.LINE_AA)

    if debug:
        show(bgr, "Input")
        show(gray, "Gray", cmap="gray")
        show(blur, f"Blur (Otsu t={t})", cmap="gray")
        show((bin_img*255).astype(np.uint8), "Otsu binary", cmap="gray")
        show((cleaned*255).astype(np.uint8), "Mask after morphology", cmap="gray")
        show(overlay, f"Overlay: {len(kept)} coins, total {euros_str(total)}")

        print("---- Detected coins details ----")
        for i, c in enumerate(kept):
            print(
                f"[{i}] r={c['radius']:.2f} area={c['area']} fam={c['family']} "
                f"dH={c['dH']:.1f} dS={c['dS']:.1f} value={c['value']:.2f}"
            )
        print("Detected coins:", len(kept))
        print("Total:", euros_str(total))

    return {
        "bgr": bgr,
        "gray": gray,
        "mask": cleaned,
        "labels": labels,
        "components": kept,
        "count": len(kept),
        "total": total,
        "overlay": overlay
    }

```

```python
import os, glob, csv
import numpy as np
import cv2

def load_ground_truth_csv(csv_path):
    """
    CSV format:
    filename,count,total
    img1.jpg,9,3.89
    """
    gt = {}
    with open(csv_path, "r", newline="", encoding="utf-8") as f:
        reader = csv.DictReader(f)
        for row in reader:
            fname = row["filename"].strip()
            gt[fname] = {
                "count": int(row["count"]),
                "total": float(row["total"])
            }
    return gt


def evaluate_folder(images_dir, gt_csv_path, debug_one=False):
    gt = load_ground_truth_csv(gt_csv_path)

    # Only keep common image extensions
    exts = ("*.jpg", "*.jpeg", "*.png", "*.bmp")
    image_paths = []
    for e in exts:
        image_paths += glob.glob(os.path.join(images_dir, e))
    image_paths = sorted(image_paths)

    results = []
    shown_one = False

    for p in image_paths:
        fname = os.path.basename(p)
        if fname not in gt:
            continue

        bgr = cv2.imread(p, cv2.IMREAD_COLOR)
        if bgr is None:
            continue

        # Run your current pipeline
        out = detect_and_count_coins(
            bgr,
            debug=(debug_one and not shown_one)
        )
        if debug_one and not shown_one:
            shown_one = True

        pred_count = out["count"]
        pred_total = out["total"]
        true_count = gt[fname]["count"]
        true_total = gt[fname]["total"]

        results.append({
            "filename": fname,
            "true_count": true_count,
            "pred_count": pred_count,
            "count_ok": int(pred_count == true_count),
            "true_total": true_total,
            "pred_total": pred_total,
            "abs_err_total": abs(pred_total - true_total),
            "count_err": abs(pred_count - true_count)
        })

    if not results:
        print("No matching images found (check folder + CSV filenames).")
        return None

    count_acc = float(np.mean([r["count_ok"] for r in results]))
    mae_total = float(np.mean([r["abs_err_total"] for r in results]))
    mae_count = float(np.mean([r["count_err"] for r in results]))

    print(f"Images evaluated: {len(results)}")
    print(f"Count accuracy (exact match): {count_acc*100:.1f}%")
    print(f"MAE(count): {mae_count:.3f} coins")
    print(f"MAE(total): {mae_total:.3f} €")

    return results
```

```python
results = evaluate_folder(
    images_dir="/content/coins_dataset",
    gt_csv_path="/content/coins_dataset/gt.csv"
)
for r in results[:10]:
    print(r["filename"], "count", r["pred_count"], "/", r["true_count"],
          "total", f"{r['pred_total']:.2f}", "/", f"{r['true_total']:.2f}",
          "abs_err", f"{r['abs_err_total']:.2f}")
```
